# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
import re
file1=[]
for i in df[0]:
  file1.append(re.split(r'\W+',i))
```

```python
file1
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
#filtered_words = [word for word in file1 if word not in stopwords.words('english')]
#print("Number of stopwords",len(file1)-len(filtered_words))
stopwords.words('english')
final1=[]
for i in file1:
  #print(i)
  filtered_words = [word for word in i if word not in stopwords.words('english')]
  final1.append(filtered_words)
```

```python
z=[]
z2=[]
for i in final1:
  z=[]
  for j in i:
    z.append(j.lower())
  z2.append(z)
```

```python
z2
```

```python
from nltk.stem import PorterStemmer 
ps = PorterStemmer() 
for w in z2:
  for k in w: 
    print(k, " : ", ps.stem(k))
```

```python
#lemetization
from nltk.stem import WordNetmy_lemmitization 
  
my_lemmitization = WordNetmy_lemmitization() 
for w in z2:
  for k in w: 
    print(my_lemmitization.lemmatize(k))
```
