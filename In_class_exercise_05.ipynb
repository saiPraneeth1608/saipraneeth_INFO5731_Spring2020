```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from textblob import Word
import nltk
from nltk.tokenize import word_tokenize 
```

```python
text = []
first_page = 'https://citeseerx.ist.psu.edu/search;jsessionid=37A0AC54277865A394D5F96748080D17?q=natural+language+processing&t=doc&sort=date'
for pg_num in range(50):
  if pg_num == 0:
    website_link = first_page
  else:
    website_link = 'https://citeseerx.ist.psu.edu/search?q=natural+language+processing&t=doc&sort=date&start=' + str (pg_num)
  page = requests.get(website_link)
  soup = BeautifulSoup(page.text, 'html.parser')
  abstracts = soup.find_all(class_='pubabstract')
  pg_num = pg_num + 10
  for abstract in abstracts:
    processed_text = abstract.text.replace('\n', '').strip()
    text.append(processed_text)
df = pd.DataFrame((text), columns =['NLP Text'])
df
```

```python
import nltk
nltk.download('stopwords')
df['Special Characters Removal'] = df['NLP Text'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', i) for i in x ))
#Removal of StopWords
stop = stopwords.words('english')
df['removed stopwords'] = df['NLP Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
print(df)
#Stemming Data
st = PorterStemmer()
df['Stemming'] = df['removed stopwords'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
#Lemmatization Data
nltk.download('wordnet')
df['Lemmatization'] = df['NLP Text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```
